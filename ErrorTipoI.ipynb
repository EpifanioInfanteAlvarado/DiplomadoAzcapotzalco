```julia
using Statistics


n = 10000    #Genero diez mil pares de muestras.
tm = 101     #Genero 101 datos en cada muestra.

t1 = 1.984   #Valor cr√≠tico superior de t para ese tama√±o de muestra.
t2 = -1.984  #Valor cr√≠tico inferior de t para ese tama√±o de muestra.


v = zeros(n)

for j in 1:n
    
   muestra1 = rand(n,tm) #Genero las primeras diez mil muestras.
   muestra2 = rand(n,tm) #Genero las segundas diez mil muestras.
    
    medias1 = mean(muestra1,dims=2) #Obtengo las primeras diez mil medias.
    medias2 = mean(muestra2,dims=2) #Obtengo las segundas diez mil medias.
     
    var1 = var(muestra1, dims=2)    #Obtengo las primeras diez mil varianzas.
    var2 = var(muestra2, dims=2)    #Obtengo las segundas diez mil varianzas.
    
    numerador = (medias1-medias2)            #Obtengo el numerador del estad√≠stico.
    denominador = sqrt.(var1/tm + var2/tm)   #Obtengo el denominador del estad√≠stico.
    t = numerador ./ denominador             #Obtengo los diez mil estad√≠sticos.
    
    
    var1 = let contador = 0
       
        for i in 1:n                        #Cuento el n√∫mero de estad√≠sticos que caen fuera del intervalo. 
            if(t[i]>t1 || t[i]<t2)
               
                #print(t[i])
                contador+=1
                
            end
            v[j]=contador
        end    
        
    end    
    
end 
    
    println("El n√∫mero de significaciones para ",  n  ," corridas es= ",v) #Muestro los estad√≠sticos
    
    media_v= mean(v,dims=1) #Se desnuda el supuesto de regularidad Estad√≠stica y en promedio se acerca a lo esperado.
    
    println(" ")

    println("El promedio de las significaciones es:", media_v) #Muestro el promedio.

    println("Observar que promedio emp√≠rico se acerca a lo esperado del error tipo I ")
```
